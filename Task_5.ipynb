```python
NUMERICAL_COLS = ['QUANTITYORDERED', 'PRICEEACH', 'SALES']
CATEGORICAL_COLS = ['STATUS', 'QTR_ID', 'PRODUCTLINE', 'DEALSIZE', 'COUNTRY', 'TERRITORY']
```

```python
print("\n\n--- 2. Initial Data Inspection (.head(), .info(), .describe()) ---")
print("\n[A] Head of the Data:")
print(df.head())

print("\n[B] Data Info (Structure, Nulls, Data Types):")
df.info()

print("\n[C] Descriptive Statistics (Numerical Columns):")
print(df[NUMERICAL_COLS].describe().T)

print("\n[D] Value Counts for Key Categorical Columns:")
for col in CATEGORICAL_COLS:
    print(f"\n--- {col} ---")
    # Display the top 10 categories and NaN counts
    print(df[col].value_counts(dropna=False).head(10))
    print(f"Total unique values: {df[col].nunique()}")

print(f"\nTotal duplicate rows: {df.duplicated().sum()}")
```

```python
print("\nUnivariate Analysis (Histograms, Boxplots, Countplots)")

# A. Numerical Data Distributions (Histograms and Boxplots)
print("\n[A] Numerical Features:")
for col in NUMERICAL_COLS:
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # Histogram/KDE
    sns.histplot(df[col].dropna(), kde=True, bins=30, ax=axes[0])
    axes[0].set_title(f'Distribution of {col}')

    # Boxplot (Outlier Detection)
    sns.boxplot(y=df[col].dropna(), ax=axes[1], color='skyblue')
    axes[1].set_title(f'Boxplot of {col}')

    plt.suptitle(f'Univariate Analysis: {col}', fontsize=16)
    plt.show()

# B. Categorical Data Frequencies (Countplots)
print("\n[B] Categorical Features:")
for col in CATEGORICAL_COLS:
    # Skip high-cardinality columns
    if df[col].nunique() > 20:
        print(f"Skipping {col} for countplot: too many unique values ({df[col].nunique()}).")
        continue

    plt.figure(figsize=(10, 6))
    sns.countplot(y=df[col], order=df[col].value_counts().index, palette='crest')
    plt.title(f'Frequency of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.show()
```

```python
# Bivariate/Multivariate Analysis 

print("Bivariate and Multivariate Analysis")

# Correlation Heatmap 
print("\n[A] Correlation Heatmap of Numerical Features:")
correlation_matrix = df[NUMERICAL_COLS].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='viridis',
            cbar=True, linewidths=.5, linecolor='black')
plt.title('Correlation Matrix of Key Numerical Features')
plt.show()


# Sales vs. Key Categorical Features 
print("\n[B] Sales Distribution by Key Categorical Features (Boxplots):")

# Use 'sales' as the target variable
target = 'SALES'

# Plot sales against key categorical features like product_line and deal_size
key_categorical = ['PRODUCTLINE', 'DEALSIZE', 'STATUS']

for col in key_categorical:
    if df[col].nunique() < 20:
        plt.figure(figsize=(12, 6))
        # Use boxplot to show median, spread, and outliers of sales across categories
        sns.boxplot(x=df[col], y=df[target], data=df, palette='hls')
        plt.title(f'{target} Distribution by {col}')
        plt.xlabel(col)
        plt.ylabel(target)
        plt.xticks(rotation=45, ha='right')
        plt.show()
    else:
        print(f"Skipping Boxplot for {col} due to high cardinality.")


# Pair Plot (
print("\n[C] Pair Plot of Sales and Top Predictors:")
# Choose 4-5 core numerical features for the pair plot
pairplot_features = ['SALES', 'QUANTITYORDERED', 'PRICEEACH', 'MSRP']
# Filter out non-existent columns just in case
pairplot_df = df[pairplot_features].dropna()

if len(pairplot_df.columns) > 1:
    # Use 'deal_size' as a hue to add multivariate context
    if 'DEALSIZE' in df.columns:
         sns.pairplot(df.dropna(subset=pairplot_features), vars=pairplot_features, hue='DEALSIZE', diag_kind='kde')
    else:
         sns.pairplot(df.dropna(subset=pairplot_features), vars=pairplot_features, diag_kind='kde')

    plt.suptitle('Pair Plot of Core Numerical Features', y=1.02)
    plt.show()
else:
    print("Not enough core numerical columns for a pair plot.")
```
